```python
import zipfile
import os

# Define the path to the zip file and the destination folder
zip_path = '/content/TrafficSignProject.zip'  # Path to your zip file
extract_path = '/content/TrafficSignProject'  # Path to the destination folder

# Unzip the folder
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Extracted contents of {zip_path} to {extract_path}")
```

```python
import os

# Define the path to the Test folder
test_folder = '/content/TrafficSignProject/TrafficSignProject/images/Test'

# Check if the 'GT-final_test.csv' file exists in the folder and remove it
csv_file = 'GT-final_test.csv'
csv_file_path = os.path.join(test_folder, csv_file)

if os.path.exists(csv_file_path):
    os.remove(csv_file_path)
    print(f"Removed the non-image file: {csv_file}")
else:
    print(f"{csv_file} not found in the Test folder.")
```

```python
import os
import shutil

# Paths
train_subfolders = '/content/TrafficSignProject/TrafficSignProject/images/Train'  # Path to Train folder with subfolders
train_flat_folder = '/content/TrafficSignProject/TrafficSignProject/images/train_flat'  # Path to destination folder

# Create train_flat folder
os.makedirs(train_flat_folder, exist_ok=True)

# Copy all images from subfolders to train_flat
for root, _, files in os.walk(train_subfolders):
    for file in files:
        if file.endswith(('.png', '.jpg', '.jpeg')):  # Adjust extensions as needed
            shutil.copy(os.path.join(root, file), os.path.join(train_flat_folder, file))

print(f"All images copied to: {train_flat_folder}")
```

```python
import os

# Define paths to the image and label folders
train_images_path = '/content/TrafficSignProject/TrafficSignProject/images/train_flat'
train_labels_path = '/content/TrafficSignProject/TrafficSignProject/labels/train_flat'

test_images_path = '/content/TrafficSignProject/TrafficSignProject/images/Test'
test_labels_path = '/content/TrafficSignProject/TrafficSignProject/labels/Test'

# List all images and labels
train_images = os.listdir(train_images_path)
train_labels = os.listdir(train_labels_path)

test_images = os.listdir(test_images_path)
test_labels = os.listdir(test_labels_path)

# Check for any missing labels in train and test sets
missing_train_labels = {img for img in train_images if img.replace('.png', '.txt') not in train_labels}
missing_test_labels = {img for img in test_images if img.replace('.png', '.txt') not in test_labels}

print(f"Missing labels for train images: {missing_train_labels}")
print(f"Missing labels for test images: {missing_test_labels}")
```

```python
from ultralytics import YOLO

# Load the YOLOv8 Nano model
model = YOLO('yolov8n.pt')  # Use yolov8n (Nano) for minimal resource usage

# Train the model
model.train(
    data='/content/TrafficSignProject/TrafficSignProject/data.yaml',  # Path to data.yaml
    epochs=10,         # Number of epochs (adjust as needed)
    imgsz=416,         # Image size (lower for faster training)
    batch=8,           # Batch size (lower for reduced GPU usage)
    device=0           # Use GPU (0), or 'cpu' if GPU is unavailable
)
```

```python
import os

# Path to search for the weights
weights_path = '/content/runs/detect/train'

# List all files in weights directory
for root, dirs, files in os.walk(weights_path):
    for file in files:
        print(os.path.join(root, file))

```

```python
from ultralytics import YOLO

# Load the trained model
model = YOLO('/content/runs/detect/train/weights/best.pt')  # Update with correct path
# Predict on a specific test image
results = model.predict(source='/content/TrafficSignProject/TrafficSignProject/images/Test/00000.png', save=True)
# Access the first result in the results list
results[0].show()  # This displays the predicted image in Colab

```
